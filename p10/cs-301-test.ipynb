```python
def get_json(path):
    url = "https://tyler.caraza-harter.com/cs301/spring19/data/capitals/" + path
    #path="C:\Users\Owner\Documents\CS 301\p10"
    if not os.path.exists(path):
        r = requests.get(url)
        r.raise_for_status()
        data = r.json()
        f=open(path, "w")
        json.dump(data, f)
        f.close()
        return data
    else:
        f=open(path, "r")
        content = json.load(f)
        f.close()
        return content
```

```python

# randall
lat1 = 43.070231
lon1 = -89.411893
#capital 
lat2 = 43.074645
lon2 = -89.384113


from math import radians, cos, sin, asin, sqrt

def haversine(lon1, lat1, lon2, lat2):
    
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2]) # descimal to radians

    # haversine formula from link  
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a)) 
    r = 3956 # radius of earth in miles 
    distance =  c * r
    return distance
```
