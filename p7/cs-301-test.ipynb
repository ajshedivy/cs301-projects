```python
import csv
fifa_file = open('Fifa18.csv', encoding='utf-8')
file_reader = csv.reader(fifa_file)
player_data = list(file_reader)
header = player_data[0]
player_data = player_data[1:]
for row in player_data:
    for idx in [2,6,7,8]:
        row[idx] = float(row[idx])
```

```python
def cell(row_idx, col_name):
    col_idx = header.index(col_name)
    val = player_data[row_idx][col_idx]
    if val == "":
        return None
    if col_name=="Age":
        val= int(val)
    elif col_name=="name":
        val=str(val)
    # optional: convert types based on column name?
    return val
```

```python
def Age():
    Alist=[]
    #best_age = 0
    row_idx=0
    while row_idx < len(player_data):
        Age=cell(row_idx, "Age")
        if Age ==None:
            row_idx+=1
            continue
        else:
            Alist.append(Age)
        row_idx+=1
    #Agelist=list(set(Alist))
    
    Alist.sort()
    x = Alist
    #print(best_age)
    return x
```

```python
def highest_Age():
    hlist = []
    alist = Age()
    row_idx = 0
    highest = alist[-1] 
    highest = int(highest)
    while row_idx<len(player_data):
        Age1 = cell(row_idx, "Age")
        if Age1 == highest:
            name = cell(row_idx, "name") 
            if str(highest) in str(Age1):
                hlist.append(name)
        row_idx+=1
        str1 = ''.join(hlist)
    return str1
```

```python
def avg_wage():
    league_counts= {}
    total_wage={}
    avg_wage={}
    for row in player_data:
        league = row[header.index("league")]
        wage= row[header.index("euro_wage")]
        if not league in league_counts:
            league_counts[league] = []
            league_counts[league].append(wage)
        else:
            league_counts[league].append(wage)
            total_wage[league]=sum(league_counts[league])
            avg_wage[league] = total_wage[league]/len(league_counts[league])
    return avg_wage
    
```

```python
def avg_age():
    league_counts= {}
    total_age={}
    avg_age={}
    for row in player_data:
        league = row[header.index("league")]
        age= row[header.index("Age")]
        if not league in league_counts:
            league_counts[league] = []
            league_counts[league].append(age)
        else:
            league_counts[league].append(age)
            total_age[league]=sum(league_counts[league])
            #print(total_wage)
            avg_age[league] = total_age[league]/len(league_counts[league])
    return avg_age
    
```
